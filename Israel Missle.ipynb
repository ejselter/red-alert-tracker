```python
import time
# Define the URL
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

url = "https://www.oref.org.il//12481-he/Pakar.aspx"

# Open the webpage
driver.get(url)

# Wait for the table to load
driver.implicitly_wait(10)  # You might need to adjust the wait time

# Set the date values using JavaScript
new_start_date = '12/08/2005'
new_end_date = '25/06/2024'

# Update start date
driver.execute_script("document.getElementById('txtDateFrom').removeAttribute('readonly');")
driver.execute_script(f"document.getElementById('txtDateFrom').value = '{new_start_date}';")

# Update end date
driver.execute_script("document.getElementById('txtDateTo').removeAttribute('readonly');")
driver.execute_script(f"document.getElementById('txtDateTo').value = '{new_end_date}';")

# Trigger change events if necessary
driver.execute_script("document.getElementById('txtDateFrom').dispatchEvent(new Event('change'));")
driver.execute_script("document.getElementById('txtDateTo').dispatchEvent(new Event('change'));")


# Wait for the table to update
time.sleep(5)  # Adjust the sleep time if necessary to allow the table to update

# Find and click all "show more" links
show_more_links = driver.find_elements(By.CLASS_NAME, "alertShowMore")
for link in show_more_links:
    driver.execute_script("arguments[0].click();", link)

# Locate all date headers and their corresponding alert tables
alert_data = []

# Find all date headers
date_headers = driver.find_elements(By.CLASS_NAME, "alertTableDate")

for date_header in date_headers:
    date = date_header.text
    # Find all alert tables following this date header
    following_siblings = date_header.find_elements(By.XPATH, "following-sibling::div[@class='alert_table']")
    
    for sibling in following_siblings:
        # Extract the category
        category = sibling.find_element(By.CLASS_NAME, "alertTableCategory").text
        
        # Extract alert details
        alert_details = sibling.find_elements(By.CLASS_NAME, "alertDetails")
        
        for alert in alert_details:
            time = alert.find_element(By.CLASS_NAME, "alertTableTime").text
            location = alert.text.replace(time, '').strip()
            alert_data.append({"date": date, "category": category, "time": time, "location": location})
        
        # Stop if the next sibling is a new date header
        next_sibling = sibling.find_element(By.XPATH, "following-sibling::*[1]")
        if next_sibling.get_attribute("class") == "alertTableDate":
            break

# Close the browser
driver.quit()

# Print the extracted data
for alert in alert_data:
    print(f"Date: {date}, Category: {alert['category']}, Time: {alert['time']}, Location: {alert['location']}")
```

* The code implemented works well but the website doesn't allow for more than 2000 results at once.
* We need to automate the process for changin the dates, switching them after we hit 2000 results
* Need to be conscious of dates overlapping - if we hit 2000 in the middle of a certain day, we need to delete those days alerts and start from there
* 

```python
import pandas as pd
from datetime import datetime, timedelta
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time as t

# Set up options for the WebDriver
chrome_options = Options()
chrome_options.add_argument("--headless")  # Ensure the browser runs in headless mode.
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Path to the chromedriver executable
webdriver_service = Service(executable_path=r"C:\Users\eitan\chromedriver-win64\chromedriver-win64\chromedriver.exe")  # Update this path as needed

# Create a new instance of the Chrome driver
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)

# Define the URL
url = "https://www.oref.org.il//12481-he/Pakar.aspx"

# Function to set date range and scrape data
def scrape_data(start_date, end_date):
    driver.get(url)
    
    # Allow some time for the page to load (if necessary)
    t.sleep(5)

    # Set the date values using JavaScript
    driver.execute_script("document.getElementById('txtDateFrom').removeAttribute('readonly');")
    driver.execute_script(f"document.getElementById('txtDateFrom').value = '{start_date}';")
    driver.execute_script("document.getElementById('txtDateTo').removeAttribute('readonly');")
    driver.execute_script(f"document.getElementById('txtDateTo').value = '{end_date}';")

    # Trigger change events if necessary
    driver.execute_script("document.getElementById('txtDateFrom').dispatchEvent(new Event('change'));")
    driver.execute_script("document.getElementById('txtDateTo').dispatchEvent(new Event('change'));")

    # Allow some time for the table to update (if necessary)
    t.sleep(5)

    # Find and click all "show more" links
    show_more_links = driver.find_elements(By.CLASS_NAME, "alertShowMore")
    for link in show_more_links:
        driver.execute_script("arguments[0].click();", link)

    # Allow some time for the hidden elements to become visible
    t.sleep(2)

    # Locate all date headers and alert tables
    alert_data = []

    # Find all elements (both date headers and alert tables)
    all_elements = driver.find_elements(By.XPATH, "//*[contains(@class, 'alertTableDate') or contains(@class, 'alert_table')]")

    current_date = None

    for element in all_elements:
        if 'alertTableDate' in element.get_attribute('class'):
            current_date = element.text
        elif 'alert_table' in element.get_attribute('class'):
            category = element.find_element(By.CLASS_NAME, "alertTableCategory").text
            alert_details = element.find_elements(By.CLASS_NAME, "alertDetails")

            for alert in alert_details:
                try:
                    time_element = alert.find_element(By.CLASS_NAME, "alertTableTime")
                    time = time_element.text
                    location = alert.text.replace(time, '').strip()
                except Exception as e:
                    time = ""
                    location = alert.text.strip()
                
                alert_data.append({"date": current_date, "category": category, "time": time, "location": location})

    return alert_data

# Function to dynamically adjust date ranges based on the number of data points
def get_alert_data_within_limits(overall_start_date, overall_end_date, max_data_points=2000):
    date_format = '%d/%m/%Y'
    all_data = []

    current_end_date = overall_end_date
    current_start_date = current_end_date - timedelta(days=30)  # Start with a month

    while current_start_date >= overall_start_date:
        print(f"Trying date range from {current_start_date.strftime(date_format)} to {current_end_date.strftime(date_format)}")
        data = scrape_data(current_start_date.strftime(date_format), current_end_date.strftime(date_format))

        if len(data) >= max_data_points:
            print(f"Reached limit with {len(data)} data points. Reducing date range.")
            current_end_date -= timedelta(days=1)
            while True:
                data = scrape_data(current_end_date.strftime(date_format), current_end_date.strftime(date_format))
                if len(data) < max_data_points:
                    break
                current_end_date -= timedelta(days=1)
        else:
            print(f"Collected {len(data)} data points from {current_start_date.strftime(date_format)} to {current_end_date.strftime(date_format)}")
            all_data.extend(data)
            current_end_date = current_start_date - timedelta(days=1)
            current_start_date = current_end_date - timedelta(days=30)
            
            if current_start_date < overall_start_date:
                current_start_date = overall_start_date

    return all_data

# Define the overall date range
overall_start_date = datetime.strptime('01/01/2023', '%d/%m/%Y')
overall_end_date = datetime.strptime('31/12/2023', '%d/%m/%Y')

# Scrape data within limits
all_alert_data = get_alert_data_within_limits(overall_start_date, overall_end_date)

# Close the browser
driver.quit()

# Create a Pandas DataFrame with all the data
df = pd.DataFrame(all_alert_data)

# Display the DataFrame
print(df)
```
